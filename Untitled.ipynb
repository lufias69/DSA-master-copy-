```python
import dsa2 as ds
```

```python
N = 160
L = 1024
p, q, g = ds.generate_params(L, N)
x, y = ds.generate_keys(g, p, q)

text = "sos"
M = str.encode(text, "ascii")
print(M)
r, s = ds.sign(M, p, q, g, x)





if ds.verify(M, r, s, p, q, g, y):
    print('All ok')
#print(M, r, s, p, q, g, y, x, sep='\n')
print("Pesan :  ", M)
print("r     :  ", r)
print("s     :  ", s)
print("p     :  ", p)
print("q     :  ", q)
print("g     :  ", g)
print("y     :  ", y)
print("x     :  ", x)
```
